```python
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = load_model('/content/drive/MyDrive/Colab Notebooks/faces/pred_model2.h5')
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/faces/unnamed.png')
```

```python
img.shape
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
image = img.reshape(1, img.shape[0], img.shape[1], 1)
```

```python
gray_img = np.array(image) / 255.0
```

```python
result = model.predict(gray_img)
```

```python
result[0]
```

```python
plt.imshow(result[0])
```

```python
plt.figure(figsize=(8,8))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('gray Image')

plt.subplot(1,2,2)
plt.imshow(result[0])
plt.title('color Image')

plt.show()
```
